# Importing The Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

# Reading The Dataset

```python
df = pd.read_csv("G:\Machine Learning Based Vehicle Performance Analyzer\car performance.csv")
df
```

# VISUALIZATION

PAIRPLOT

```python
sns.pairplot(df,diag_kind='kde')
plt.show()
```

HEATMAP

```python
sns.heatmap(df.corr(),annot=True,linecolor ='black', linewidths = 1)
fig=plt.gcf()
fig.set_size_inches(5,5)
```

SEPERATING THE DATA

```python
x=df[['cylinders','displacement','horsepower','weight','model year','origin']].values #INDEPENDENT VARIABLE
x
```

```python
y=df.iloc[:,0:1].values #DEPENDENT VARIABLE
y
```

# Splitting The Dataset Into Dependent And Independent Variable

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

Testing the Algorithms

1.Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred1=reg.predict(x_test)
y_pred1
```

```python
ax1 = sns.distplot(df['mpg'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred1, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for mpg')
plt.xlabel('mpg')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

Calculating Metrics

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred1)
```

```python
mean_squared_error(y_test,y_pred1)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred1))
```
